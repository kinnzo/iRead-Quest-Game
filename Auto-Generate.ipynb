```python
import pandas as pd
import xml.etree.ElementTree as ET
from xml.dom import minidom as minidom
```

```python
def addrow(root,row):
    room = ET.Element("object", name=row[0])
    thetail = "\n"
    room.tail = thetail
    in4 = ET.Element("inherit", name="editor_room")
    in4.tail = thetail
    des = ET.Element("description")
    des.tail=thetail
    ext = ET.Element("exit", to=row[4])
    ext.tail = thetail
    in5 = ET.Element("inherit", name=row[5]+"direction")
    in5.tail = thetail
    ext.append(in5)
    room.append(in4)
    room.append(des)
    room.append(ext)
    key = "./object[@name=\""+row[0]+"\"]"
    rflag = 0
    for rooms in root.findall(key):
        print(rooms.attrib)
        print(rooms.tag)
        tags = rooms.getchildren()
        for t in tags:
            print(t.attrib)
        if tags[0].attrib['name'] == "editor_room":
            room = rooms
            root.remove(room)
    des = room.find("description")
    room.remove(des)
    if des.text is not None:
        des.text = des.text + " \n {img:"+row[3]+"}"
    else:
        des.text = "Your challenge is to find: \n <br /> {img:"+row[3]+"}"
    room.append(des)
    obj = ET.Element("object", name=row[1])
    obj.tail = thetail
    in1 = ET.Element("inherit", name="editor_object")
    in1.tail = thetail
    in2 = ET.Element("inherit", name="surface")
    in2.tail = thetail
    fc = ET.Element("feature_container")
    fc.tail = thetail
    hc = ET.Element("hidechildren")
    hc.tail = thetail
    lc = ET.Element("listchildren")
    lc.tail = thetail
    subobj = ET.Element("object", name=row[2])
    subobj.tail = thetail
    in3 = ET.Element("inherit", name="editor_object")
    in3.tail = thetail
    tk = ET.Element("take")
    tk.tail = thetail
    tkmsg = ET.Element("takemsg")
    tkmsg.tail = thetail
    tkmsg.text = "You picked up the "+row[2]+"."
    subobj.append(in3)
    subobj.append(tk)
    subobj.append(tkmsg)
    obj.append(in1)
    obj.append(in2)
    obj.append(fc)
    obj.append(hc)
    obj.append(lc)
    obj.append(subobj)
    room.append(obj)
    root.append(room)
    tree.write("ReQuest.aslx")
```

```python
db = pd.read_csv('myedit.csv')
tree = ET.parse('ReQuest.aslx')
root = tree.getroot()
for i in range(len(db)):
    addrow(root,db.iloc[i])
```
